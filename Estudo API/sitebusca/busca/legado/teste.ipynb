```python
import googlemaps
import json
import requests
```

```python
cep = 58051200
```

```python
mapas =requests.get(f'https://cep.awesomeapi.com.br/json/{cep}')
endereco_dic = mapas.json()
print(endereco_dic)

lat = endereco_dic['lat']
lng = endereco_dic['lng']
location = (lat, lng)

```

```python
API_KEY = ''
map_client = googlemaps.Client(API_KEY)
```

```python
print(dir(map_client))
```

```python
query1 = map_client.places_nearby(location=location, radius=1000,type='bar', keyword='bar')
print(query1)
```
